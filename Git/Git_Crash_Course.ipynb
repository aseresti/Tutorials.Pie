# Version Control System
VCS basically is taking snapshots of the project at any wanted time. Later on, the snapshots are available to look at them and their differences from the previous ones.

### Why do we need VCS?
 - Backup/Archive
 - Versioning/ History
 - Undo Changes
 - Collaboration/Teamwork
 - Blame! :D
 - Isolation of Changes
 - Experimentation
 - Context Switching: change branches
 - Code Review

### Centeralized vs. Distributed VCS
##### Centeralized VCS
In the centeralized VCS there is a single "central" copy of the project on some server, and programmers will commit changes to the central copy. The copy of the project is not on the hard drive manually. Also, all the time a connection to the central server is needed.

##### Distributed VCS
There may be or not a central server, and each developer clones a copy of the repository and has a full history of the project on their own hard drive. In this method most operations are local and the central server is not required.

##### Git is a distributed VCS.

### Git installation
On MAC you can use ```brew install git```
then the git command should be activated on your terminal. Next step is to create a git-hub account.
#### Git vs. GitHub
Git is the version control system, and github is the platform on which the central repository of the project is saved. There are several other centralized systems that can replace github such as Google Cloud, and Bitbucket.

### Key Concepts
First, on Git, we need to initiate a project in a directory, so that every versions of the project would be managed on that dir. on your terminal type
```
git init
```
This directory is your project working directory. Should you type ```ls -a``` on terminal you will see a .git folder in the directory which is the git repository and different versions of the projects gets preserved inside the .git dir. .git basically is a local repo and pushing it on the git-hub makes it a remote repo.
Inbetween the working directory and repository (.git folder) there is a staging area aka Git index. This holdes the area for queuing up changes for the next commit. Therefore, there are 4 steps between the project working directory to pushing it on the central server:

 - Working Directory
 - Staging Area
 - Git Repository or the Commit History
 - Remote Repo



### Cloning an Existing Repo: ```git clone```
While ```git init``` initiates a new repository, ```git clone``` is the command lne to target an existing repo on the central server, github, and save a copy of that on your local dir.
There are several methods of cloning:

 - HTTPS
 - SSH
 - GitHub CLI

To clone using https you need to copy the github specific url of the repo and in the project directory on your terminal type ```git cline https://github.com/link/to/remote/repo.git``` 
ssh is a different protocol that is more secure and needs a SSH-Key to tunnel to the central server. 

You may create the git repo on github. Add the name and description of the repo and you can also add 3 options to add to your repo:
 - README: this is a file ```README.md``` containing the help description on how to use the tools and codes on the repo.
 - .gitignore
 -Lisence: you may add a lisence to your repository for the copy-right and/or liability issues.

On github we have two different type of repository: research, and product.
On the research repository, our focus is on the research and the proof of the concept. research repo contains a folder named src. On the main folder of the project, files such as .gitignore, LICENCE, README.md, requirement.txt, etc are existed. The src folder contins the source code of the project. The requirement.txt contains the required libraries and their version that is used in the project.

## Git Command Lines after clonning a project

#### ```git status```
The first command we need to use is the ```git status``` command to check the branch we are in and if the local repo is uptodated to the latest version of the remote repo. It also gives a status of the modified and untracked files on the working directory.
#### ```git add```
This command line adds files on the working directory to the staging area.
You can use ```git add YOURFILENAME.ext``` or ```git add *```



#### ```git commit```
This command line adds the files from the staging area to the remote repository. you also need to write a message while commiting to the remote repo. ```git commit -m "your message"```



#### ```git log```
This command line shows the history of the commits. Every commit is identified with a hash id which is shown with it and then it shows the location of the branch with regard to the remote repo using Head -> main for local and origin/main, origin/Head for the remote repo. It also shows the author of the commit.

#### ```git push```
This command line pushes the commits on the local directory to the remote repo.

#### ```git pull```
This command updates the local repository from the remote repository.

#### ```get diff```
This command line shows the specific changes made on a single file from the previous version of that.
```git diff name_of_the_file.ext```
